```python
import pandas as pd
import plotly.express as px

# Download and load the dataset
url = 'https://www.camara.leg.br/cotas/Ano-2023.csv.zip'
politics_df = pd.read_csv(url, compression='zip', encoding='UTF-8', sep=';')
```

```python
gastos_gas_df = politics_df[politics_df['txtDescricao'].str.contains('COMBUSTÍVEIS E LUBRIFICANTES.')].reset_index(drop=True)
```

```python
# Remove the rows where txNomeParlamentar is null or contains LID.

gastos_gas_df = gastos_gas_df[gastos_gas_df['txNomeParlamentar'].notnull()].reset_index(drop=True)

gastos_gas_df = gastos_gas_df[~gastos_gas_df['txNomeParlamentar'].str.contains('LID')].reset_index(drop=True)


```

```python
#keep only columns txtNomeParlamentar, sgUF, txtDescricao, sgPartido, txtCNPJCPF and vlrLiquido

gastos_gas_df = gastos_gas_df[['txNomeParlamentar', 'sgUF', 'txtDescricao', 'sgPartido', 'txtCNPJCPF', 'vlrLiquido']]
```

```python
gastos_gas_df.head()
```

```python
#create a new df called gastos_gas_totals and group by txNomeParlamentar and summing all expenses by txNomeParlamentar. Keep sgPartido and sgUF

gastos_gas_totals = gastos_gas_df.groupby('txNomeParlamentar').sum().reset_index()
```

```python
gastos_gas_totals.sort_values(by='vlrLiquido', ascending=False)
```

```python
#Ploting using plotly express the top 10 politicians with the highest expenses in gas.

fig = px.bar(gastos_gas_totals.head(10), x='txNomeParlamentar', y='vlrLiquido', color='txNomeParlamentar', title='Top 10 politicians with the highest expenses in gas')

fig.show()
```

```python
#calculates the average expense in gas by 12 months adding a new column gasto_mensal

gastos_gas_totals['gasto_mensal'] = gastos_gas_totals['vlrLiquido']/12

gastos_gas_totals.sort_values(by='gasto_mensal', ascending=False)



```
